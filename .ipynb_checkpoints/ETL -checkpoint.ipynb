```python
import pandas as pd
covid_data=pd.read_json(statistics_json)
covid_data.head()
```

```python
# Adding new columns to separate the information in the columns cases, deaths and tests.

covid_data["total cases"]=covid_data.loc[:, "cases"].apply(lambda x: x.get("total"))
covid_data["new cases"]=covid_data.loc[:,"cases"].apply(lambda x:x.get("new"))
covid_data["total deaths"]=covid_data.loc[:, "deaths"].apply(lambda x: x.get("total"))
covid_data["new deaths"]=covid_data.loc[:,"deaths"].apply(lambda x:x.get("new"))
covid_data["total tests"]=covid_data.loc[:, "tests"].apply(lambda x: x.get("total"))
```

```python
covid_data.head()
```

```python
# Dropping the columns with extra information.

covid_data.drop(['cases','deaths','tests'],axis=1,inplace=True)
```

```python
# Filling the missing values in multiple columns

covid_data_updated=covid_data.fillna({'new cases':0,'total deaths':0,'new deaths':0,'total tests':0,'population':0})
```

```python
# Standardizing column formats

covid_data_updated=covid_data_updated.astype(
    {"population":"int64",
     "total deaths":"int64",
     "total tests":"int64",
     "day":"datetime64[ns]",
     "time":"datetime64[ns]",
     "new cases":"int16",
     "new deaths":"int16"})
```
